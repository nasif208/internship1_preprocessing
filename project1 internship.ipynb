```python
merged_data=merged_data.drop(['accept_gps_time','delivery_gps_time'],axis=1)
```

```python
merged_data['accept_time']=pd.to_datetime(merged_data['accept_time'],format="%m-%d  %H:%M:%S",errors='coerce')
merged_data['delivery_time']=pd.to_datetime(merged_data['delivery_time'],format="%m-%d  %H:%M:%S",errors='coerce')
```

```python
merged_data
```

```python
merged_data[['lng', 'lat']] = merged_data[['lng', 'lat']].astype(float)
```

```python
merged_data['aoi_type'] = merged_data['aoi_type'].astype('category')
```

```python
merged_data.info()
```

```python
diffrent_lng= (merged_data['lat'].between(-90, 90)) & (merged_data['lng'].between(-180, 180))
merged_data = merged_data[diffrent_lng]
```

```python
merged_data['delivery_duration'] = (merged_data['delivery_time'] - merged_data['accept_time']).dt.total_seconds() / 60
```

```python
from scipy.stats import zscore
```

```python
z_scores = zscore(merged_data['delivery_duration'])
merged_data= merged_data[(z_scores > -3) & (z_scores < 3)]
```

```python
z_scores
```

```python
colums= ['lng', 'lat', 'delivery_duration']
scaler = StandardScaler()
merged_data[colums] = scaler.fit_transform(merged_data[colums])
```
